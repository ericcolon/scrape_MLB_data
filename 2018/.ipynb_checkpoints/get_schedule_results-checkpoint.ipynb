```python
import pandas as pd

def _get_opposing_pitcher(df):
    
    wins = df[(df['W/L'] == 'W') | (df['W/L'] == 'W-wo')]
    loss = df[(df['W/L'] == 'L') | (df['W/L'] == 'L-wo')]
    
    wins['opp_pitcher'] = wins['Loss']
    loss['opp_pitcher'] = loss['Win']
    
    new = pd.concat([wins,loss])
    return new

#Append above column to hitter data
#Will need once we scrape more hitter data
def append_opposing_pitcher(hitter_data,schedule_df):
    #Reformat hitter dates
    hitter_data['Game Date'] = hitter_data['Game Date'].apply(lambda x: x.replace('\\r\\n','').strip())
    hitter_data['Game Date'] = pd.to_datetime(hitter_data['Game Date'],format="%m/%d/%Y",errors='coerce')
    schedule_df['Date'] = pd.to_datetime(schedule_df['Date'],format="%m/%d/%Y",errors='coerce')
    
    df1 = pd.merge(hitter_data,schedule_df[['Date','Opp','opp_pitcher','Gm#','Team']],how='left',
                   left_on=['Team','Game Date'],right_on=['Team','Date'])
    df1.dropna(inplace=True)
    return df1
```

```python
all_dfs = []
for f in os.listdir('_data/schedules/original_2017'):
    print 'Working on %s...' %f
    df = pd.read_csv('_data/schedules/original_2017' +'/'+ f)
    df1 = _get_opposing_pitcher(df)
    
    sch = df1[['Date','Opp','opp_pitcher','Gm#']]
    sch['Team'] = baseballRef_to_baseballMus.get(f)
    
    
    all_dfs.append(sch)
    #df1.to_csv(TRANS_SCHEDULES + '\\'+ f.replace('.txt','.csv'))
```
